## Discretizing functions with finite elements

As we saw in class, we can discretize a function inside of an element by 

$u^e(x) = \sum N_i^e(x)u_i^e$

Or the derivative of the function by 

$\frac{du^e(x)}{dx} = \sum \frac{dN_i^e(x)}{dx} u_i^e$

And this is true whether we have linear shape functions, quadratic shape functions, etc... Furthermore, this holds in 1D, 2D or 3D. 

In this notebook we define linear and quadratic shape functions using the concept of Lagrange polynomials in 1D. We build functions directly in terms of the physical coordinate $x$, but *warning*, in practice we will use the isoparametric map which is a concept covered in the next lecture. 

```python
import numpy as np
import sympy as sp
import matplotlib
import random
import matplotlib.pyplot as plt
%matplotlib inline 
```

Consider the mesh in the figure, made out of linear elements in 1D. Each element is made out of two nodes. 

<img src="images/mesh_linear_1D.png" alt="mesh example in 1D with 2-node linear elements" style="width: 400px;"/>

The mesh is defined by two arrays, one containing the nodal positions and one containining the *connectivity* information, i.e. which nodes make up each element. Note that the nodes not need be in any order, but for the elements we do need to make sure we list the left node first and the right node of the element second. See below. 

```python
n_ele = 4
n_nod = 5
elements = np.array([[0,2],[2,3],[3,4],[4,1]])
node_X = np.array([0.,1.,0.25,0.5,0.75])
```

In order to interpolate inside of the domain of each element and eventually in the whole domain, we need to define the shape functions. The Lagrange polynomial approach guarantees that the shape function is $1$ at the desired node and $0$ at the other nodes:

$$N_1(x) = \frac{x - x_2^e}{x_1^e-x_2^e}$$

is such that $N_1(x_1^e)=1$ and $N_1(x_2^e)=0$. Similarly, for the second shape function we have 

$$N_2(x) = \frac{x - x_1^e}{x_2^e-x_1^e}$$

is such that $N_2(x_1^e)=0$ and $N_2(x_2^e)=1$. 

```python
## Define shape function
def N1(x,x1e,x2e):
    le = x2e-x1e
    return (1./le)*(x2e-x)

def N2(x,x1e,x2e):
    le = x2e-x1e
    return (1./le)*(x-x1e)

## Define gradients of shape functions 
def dN1dx(x,x1e,x2e):
    le = x2e-x1e
    return -1./le

def dN2dx(x,x1e,x2e):
    le = x2e-x1e
    return 1./le
```

Let's just plot these functions, for an arbitrary element with two random nodes: 

$$ x_1^e = 5$$
$$ x_2^e = 11$$ 

Plot the shape functions 

```python
font = {'family' : 'Arial',
        'size'   : 18}
matplotlib.rc('font', **font)
```

```python
## create a bunch of x points to plot 
x1e = 5
x2e = 11
x_ls = np.linspace(x1e,x2e,50)
plt.plot(x_ls,N1(x_ls,x1e,x2e),label=r'$N_1(x)$',color='royalblue')
plt.plot(x_ls,N2(x_ls,x1e,x2e),label=r'$N_2(x)$',color='hotpink')
plt.xlabel('x')
plt.legend(loc=(0.33,0.68))
```

Now let's actually interpolate a function ove the entire domain. As discussed in the lecture video, the function is parameterized by the nodal values. These are basically like dials, while the shape functions allow us to fill in the function inside of each element. Without the shape functions, all we have is a bunch of scalar values $u_i$ as shown in the figure

<img src="images/mesh_linear_1D_u.png" alt="mesh example in 1D with 2-node linear elements" style="width: 400px;"/>

But with the shape functions we are going to interpolate, as mentioned at the top of the notebook, for the function we have 

$u^e(x) = \sum N_i^e(x)u_i^e$

And the derivative of the function is given by 

$\frac{du^e(x)}{dx} = \sum \frac{dN_i^e(x)}{dx} u_i^e$

Note that the interpolation is defined inside of each element. In order to get the plot of the entire function over the mesh we need to loop over elements and plot the function and derivative for each element. 

```python
# Fill in an array with the nodal values of u, maybe some random number here
node_u = np.random.rand(5)
```

```python
## plot a function over the domain by plotting in each element 

fig, (ax1,ax2) = plt.subplots(1, 2, figsize=(10,4))

# Loop over elements 
for i in range(n_ele):
    
    # Read the nodes that make up this element 'i'
    n1 = elements[i,0] #note I don't need the '-1' because I numbered nodes from '0'
    n2 = elements[i,1]
    
    # Beginning and end 'x' coordinates of this element 
    x1e = node_X[n1]
    x2e = node_X[n2]
    
    # Get the nodal values of the function 'u'
    u1e = node_u[n1]
    u2e = node_u[n2]
    
    # Eval the shape function evaluation for multiple x in this element
    x_vec = np.linspace(x1e,x2e,50)
    
    # use the shape functions to interpolate from the nodal u values to all the 'x'
    u_vec = N1(x_vec,x1e,x2e)*u1e + N2(x_vec,x1e,x2e)*u2e
    
    # gradient is going to be constant in this case
    dudx = dN1dx(x_vec,x1e,x2e)*u1e + dN2dx(x_vec,x1e,x2e)*u2e
    dudx_vec = dudx*np.ones(50)
    
    # plot function u
    ax1.plot(x_vec,u_vec)
    
    # plot the gradient
    ax2.plot(x_vec,dudx_vec)
ax1.set_xlabel('x')
ax2.set_xlabel('x')
ax1.set_ylabel('u(x)')
ax2.set_ylabel('du(x)/dx')
plt.tight_layout()
```
